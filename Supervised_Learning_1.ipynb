#### I. GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

We will experiment with KNN hyperparameters and GridSearchCV algorithm's CV value to arrive at the best performance

```python
#setting the Hyperparameters range
n_neighbors = list(np.arange(1,20,2)) #Initializing odd number of neighbors
weights = ['uniform','distance']
metric = ['minkowski','euclidean','manhattan']

#Initializing the hyperparameters dictionary
hyperparameters = dict(n_neighbors=n_neighbors, weights=weights, metric=metric)

KNN_2 = KNeighborsClassifier()

scores_gs = []
#Trying out GridSearchCV with various CV values
for i in range(3,14):
    clf = GridSearchCV(KNN_2, hyperparameters, cv=i, n_jobs=-1)
    #Fitting the model
    best_model = clf.fit(X_train, y_train)
    print('Best score with cv = {0} is {1}'.format(i,best_model.best_score_))
    scores_gs.append(best_model.best_score_);
```

```python
plt.plot(range(3,14),scores_gs)
```

**So we see the best score is reached with cross-validation value 11**

Building model using GridSearchCV with CV value 11

```python
clf = GridSearchCV(KNN_2, hyperparameters, cv=11, n_jobs=-1)
#Fitting the model
best_model = clf.fit(X_train, y_train)
```

```python
best_model.best_estimator_.get_params()
```

```python
best_model.best_score_
```

```python
print('The parameters of the new model are: ',best_model.best_params_)
```

**So we see that the best selected parameters of the new model are:**

Distance Metric: Minkowski

p (Power): 2

Weight: Uniform
